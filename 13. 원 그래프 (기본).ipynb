# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 고딕
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.f%%', startangle=90, 
       counterclock=False
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
# explod = [0.2, 0.1, 0, 0, 0, 0]
explod = [0.05] * 6

plt.pie(values, labels=labels, autopct='%.f%%', startangle=90, 
       counterclock=False, explode=explod)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explod)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explod)
plt.title('언어별 선호도')
plt.legend(loc=(1.2, 0.3), title = '언어별 선호도')
plt.show()
```
